#  平均の区間推定

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()

df = pd.read_csv('kaggle_income.csv', encoding="ISO-8859-1")
df.columns
```

```python
sns.displot(df['Mean'])
```

```python
np.mean(df['Mean'])
```

1.母集団から無作為抽出した標本を作成する

```python
n = 1000
sample_df = df.sample(n=n)
sns.displot(sample_df['Mean'])
```

2.標本から推定量を計算する

```python
np.mean(sample_df['Mean'])
```

3.信頼区間を決める

4.推定量の標本分布を考える

```python
from scipy import stats
sample_mean=np.mean(sample_df['Mean'])
sample_var = stats.tvar(sample_df['Mean'])
stats.norm.interval(alpha=0.95, loc=sample_mean, scale=np.sqrt(sample_var/n))
```

# t分布

```python
x = np.linspace(-3, 3, 100)
for df in range(1, 6):
    t = stats.t.pdf(x, df)
    plt.plot(x, t, label='df={}'.format(df))
plt.legend()
```

#### 標準正規分布と比較する

```python
x = np.linspace(-3, 3, 100)
z = stats.norm.pdf(x, loc=0, scale=1)
for df in range(1, 10, 4):
    t = stats.t.pdf(x, df)
    plt.plot(x, t, label='df={}'.format(df))
plt.plot(x, z, label='std norm', linewidth=3)
plt.legend()
```

## 95%信頼区間

```python
x = np.linspace(-3, 3, 100)
left, right = stats.norm.interval(0.95, loc=0, scale=1)
z = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, z)
plt.axvline(left, c='r')
plt.axvline(right, c='r')
```

```python
x = np.linspace(-5, 5, 100)
left, right = stats.t.interval(0.95, df=2)
t = stats.t.pdf(x, df=2)
plt.plot(x, t)
plt.axvline(left, c='r')
plt.axvline(right, c='r')
```

## t分布を使って平均値の区間推定

```python
df = pd.read_csv('kaggle_income.csv', encoding="ISO-8859-1")
```

```python
print('Numpy:{}'.format(np.mean(df['Mean'])))
print('Pandas:{}'.format(df['Mean'].mean()))
```

```python
n = 1000
sample_df = df.sample(n=n)
print(sample_df['Mean'].mean())
```

```python
sample_mean = sample_df['Mean'].mean()
sample_std = sample_df['Mean'].std()
stats.t.interval(0.95, loc=sample_mean, scale=sample_std/np.sqrt(n), df=n-1)
```

```python
stats.norm.interval(0.95, loc=sample_mean, scale=sample_std/np.sqrt(n))
```

# t分布

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

%matplotlib inline
sns.set()

df = pd.read_csv('titanic_train.csv')
df.head()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
ages1 = df[df['Survived']==1]['Age']
ages2 = df[df['Survived']==0]['Age']
```

```python
print(ages1.mean(), ages2.mean())
```

```python
t, p = stats.ttest_ind(ages1, ages2, alternative='less')
print('t値：{}', t)
print('p値：{}', p)
```

```python
x = np.linspace(-4, 4, 100)
y = stats.t.pdf(x, df=len(ages1)+len(ages2)-2)
plt.axvline(t, c='r')
plt.plot(x, y)
```

```python
stats.t.cdf(t, df=len(ages1)+len(ages2)-2)
```

# F検定

```python
from scipy import stats

x = np.linspace(-1, 3, 100)
for dfn in range(1, 12, 5):
    for dfd in range(1, 12, 5):
        y = stats.f.pdf(x, dfn=dfn, dfd=dfd)
        plt.plot(x, y, label=f'dfn:{dfn}, dfd:{dfd}')
plt.legend()
```

```python
n1 = len(ages1)
n2 = len(ages2)
var1 = stats.tvar(ages1)
var2 = stats.tvar(ages2)
dfn = n2 - 1
dfd = n1 - 1
```

```python
f = var2 / var1
print(f)
```

```python
stats.f.cdf(f, dfn=dfn, dfd=dfd)
```

```python
p = 1 - stats.f.cdf(f, dfn=dfn, dfd=dfd)
p
```
